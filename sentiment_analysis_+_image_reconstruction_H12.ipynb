```python
from PIL import Image
from diffusers import StableDiffusionImg2ImgPipeline
import torch

# Load the Stable Diffusion model (ensure you have a GPU for better performance)
pipe = StableDiffusionImg2ImgPipeline.from_pretrained("runwayml/stable-diffusion-v1-5").to("cuda")

def generate_images(input_image_path, prompt, iterations=2, strength=0.75, output_prefix="sufes"):
    """
    Generates images iteratively using the same prompt.

    Parameters:
    - input_image_path (str): Path to the initial image.
    - prompt (str): Text prompt for image generation.
    - iterations (int): Number of images to generate in sequence.
    - strength (float): Controls how much the output resembles the input image.
    - output_prefix (str): Prefix for saved images.

    Returns:
    - List of paths to the generated images.
    """
    output_paths = []
    current_image = Image.open(input_image_path).convert("RGB").resize((512, 512))

    for i in range(1, iterations + 1):
        # Generate a new image
        result = pipe(prompt=prompt, image=current_image, strength=strength).images[0]

        # Save the image with an iteration label
        output_name = f"{output_prefix}_iter{i}.png"
        result.save(output_name)
        output_paths.append(output_name)

        print(f"Generated Image {i}: {output_name}")

        # Use the generated image as input for the next iteration
        current_image = result

    return output_paths


# Example usage:
prompt_text = "A detailed reconstruction of a Roman archaeological site in Sufes, Tunisia with realistic textures and natural lighting."
generated_images = generate_images("reconstructed_image1.png", prompt_text, iterations=2)

print("Generated images:", generated_images)
```

```python
import fitz  # PyMuPDF
import os
from PIL import Image
import pandas as pd

# Define the file path
pdf_path = "/content/Synthèse volume I (3).pdf"
output_dir = "D:/3IA/s2/hack/extracted_images"

# Create directory for extracted images if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Open the PDF file
doc = fitz.open(pdf_path)
image_data = []

# Loop through each page
for page_number in range(len(doc)):
    page = doc[page_number]
    text = page.get_text("text")  # Extract text from the page

    # Extract images from the page
    for img_index, img in enumerate(page.get_images(full=True)):
        xref = img[0]
        base_image = doc.extract_image(xref)
        image_bytes = base_image["image"]
        img_ext = base_image["ext"]

        # Save the image
        image_filename = f"page_{page_number+1}_img_{img_index+1}.{img_ext}"
        image_path = os.path.join(output_dir, image_filename)

        with open(image_path, "wb") as img_file:
            img_file.write(image_bytes)

        # Store image metadata and related text
        image_data.append({
            "page": page_number + 1,
            "image_path": image_path,
            "context": text[:500]  # Capture first 500 characters as context
        })



# Save the extracted data into a CSV for easy viewing
df_images = pd.DataFrame(image_data)
csv_output_path = "extracted_images_with_context.csv"
df_images.to_csv(csv_output_path, index=False)

print(f"Extraction complete! Images saved in '{output_dir}' and data saved as '{csv_output_path}'.")

```

```python
import re
import os

# Load CSV
csv_path = "extracted_images_with_context.csv"
df = pd.read_csv(csv_path)

# Clean image paths (Keep just file names)
df["image_path"] = df["image_path"].apply(lambda x: os.path.basename(x))

# Remove unwanted characters from context
def clean_text(text):
    text = re.sub(r"\n", " ", text)  # Remove newlines
    text = re.sub(r"9([a-zA-Z])", r"\1", text)  # Fix encoding issues
    return text.strip()

df["context"] = df["context"].apply(clean_text)

# Save the cleaned CSV
clean_csv_path = "cleaned_extracted_images_with_context.csv"
df.to_csv(clean_csv_path, index=False)

print(f"Cleaned file saved as '{clean_csv_path}'.")
```
